```python
import numpy as np
```

```python
t = np.arange(0, 1.2, 0.2)
analytic = np.exp(t)

explicit_euler = np.zeros(np.size(t))
explicit_euler[0] = 1

implicit_euler = np.zeros(np.size(t))
implicit_euler[0] = 1


for n in range(1, np.size(t)):
    explicit_euler[n] = explicit_euler[n-1] + explicit_euler[n-1] * (t[n] - t[n-1])
    implicit_euler[n] = implicit_euler[n-1]/(1-(t[n]-t[n-1]))    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t, analytic)
plt.plot(t, explicit_euler)
plt.plot(t, implicit_euler)
```

```python
np.savetxt('data/euler_analytic.txt', np.matrix([t, analytic]).T)
np.savetxt('data/euler_explicit.txt', np.matrix([t, explicit_euler]).T)
np.savetxt('data/euler_implicit.txt', np.matrix([t, implicit_euler]).T)
```
